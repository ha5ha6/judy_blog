**vector**: arrow starting from origin, with directions of magnitudes of two axes

$\begin{align}
\vec{v} &= \begin{bmatrix}
       1 \\
       2 \\
     \end{bmatrix}
\end{align}$

**"go to the direction of 1 step at x-axis, and 2 steps at y-axis"**

**"if you are dealing with a collection of vectors, it's convenient to think of them all as points"**

**vector operations**:

1. $\vec{v}+\vec{w}$

2. $c\vec{v}$

**unit vectors**:

$\begin{align}
\vec{i} &= \begin{bmatrix}
       1 \\
       0 \\
     \end{bmatrix}
\end{align}$, $\begin{align}
\vec{j} &= \begin{bmatrix}
       0 \\
       1 \\
     \end{bmatrix}
\end{align}$

e.g. $\begin{align}
\begin{bmatrix}
       3 \\
       -2 \\
     \end{bmatrix}
\end{align}$ can be interpreted as scale $\vec{i}$ with 3 units, and $\vec{j}$ with -2 units

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))
plt.arrow(0,0,1,2,color='r',head_width=0.2,length_includes_head=True,label='[1,2]')
plt.arrow(0,0,1,0,color='b',head_width=0.2,length_includes_head=True,label='i=[1,0]')
plt.arrow(0,0,0,1,color='g',head_width=0.2,length_includes_head=True,label='j=[0,1]')
plt.plot(0,0,'ro',markersize=12,label='origin')
plt.legend(fontsize=14)
plt.xlim([-5,5])
plt.ylim([-5,5])
plt.xticks(np.arange(-5,5))
plt.yticks(np.arange(-5,5))
plt.grid()
plt.gca().set_aspect("equal")
```

**linear combination**: $a\vec{v}+b\vec{w}$

**span** of $\vec{v}$ and $\vec{w}$: the set of all their linear combinations

e.g. span of $\begin{align}
\vec{v} &= \begin{bmatrix}
       1 \\
       2 \\
     \end{bmatrix}
\end{align}$, $\begin{align}
\vec{w} &= \begin{bmatrix}
       -2 \\
       3 \\
     \end{bmatrix}
\end{align}$

**the span of most pairs of vectors ends up being the entire infinite sheet of a 2d space or a line**

**the span of two vectors in a 3d space is a flat sheet**

**linearly dependent**: one of the vectors can be expressed as a linear combination of the other, since it's already in the span of the other

$\vec{w}=a\vec{v}$

**linearly independent**:

$\vec{w} \neq a\vec{v}$

**basis**: of a vector space is a set of linearly independent vectors that span the full space

```python
plt.figure(figsize=(6,6))
v=np.array([1,2])
w=np.array([-2,3])
ab=np.random.uniform(-2,2,(100,2))

for i in range(ab.shape[0]):
    lc=ab[i][0]*v+ab[i][1]*w
    plt.arrow(0,0,lc[0],lc[1],color='r',head_width=0.2,length_includes_head=True)

plt.xlim([-10,10])
plt.ylim([-10,10])
plt.xticks(np.arange(-10,10))
plt.yticks(np.arange(-10,10))
plt.grid()
```

**linear transformation**

**transformation**: vector -> another vector, or matrix -> another matrix

(like a function to the variable)

**linear transformation**:

1. lines remain lines (no curves)

2. origin remains fixed

in general **keeping grid lines parallel and evenly spaced**

e.g. suppose $\begin{align}
\vec{v} &= \begin{bmatrix}
       -1 \\
       2 \\
     \end{bmatrix}
\end{align}=-1 \vec{i}+2 \vec{j}$

$\vec{v}_{\text{transformed}} =-1 \vec{i}_{\text{transformed}}+2 \vec{j}_{\text{transformed}}$

suppose $\begin{align}
\vec{i}_{\text{transformed}} &= \begin{bmatrix}
       1 \\
       -2 \\
     \end{bmatrix}
\end{align}$, $\begin{align}
\vec{j}_{\text{transformed}} &= \begin{bmatrix}
       3 \\
       0 \\
     \end{bmatrix}
\end{align}$, where $\vec{i}$ and $\vec{j}$ landed after transformation

$\begin{align}
\vec{v}_{\text{transformed}} = -1 \begin{bmatrix}
       1 \\
       -2 \\
     \end{bmatrix}+2 \begin{bmatrix}
       3 \\
       0 \\
     \end{bmatrix}=\begin{bmatrix}
       5 \\
       2 \\
     \end{bmatrix}
\end{align}$

**in a matrix formation**:

suppose $\begin{align}
\vec{i} \rightarrow \begin{bmatrix}
       1 \\
       -2 \\
     \end{bmatrix}
\end{align}$, $\begin{align}
\vec{j} \rightarrow \begin{bmatrix}
       3 \\
       0 \\
     \end{bmatrix}
\end{align}$,

$\begin{align}
\begin{bmatrix}
       x \\
       y \\
     \end{bmatrix}
\rightarrow x \begin{bmatrix}
       1 \\
       -2 \\
     \end{bmatrix} + y \begin{bmatrix}
       3 \\
       0 \\
     \end{bmatrix} = \begin{bmatrix}
       1x+3y \\
       -2x+0y \\
     \end{bmatrix}
\end{align}$

=>$\begin{align}
\begin{bmatrix}
       1 & 3 \\
       -2 & 0 \\
     \end{bmatrix}\begin{bmatrix}
       x \\
       y \\
     \end{bmatrix}
\end{align}$

in general

$\begin{align}
\begin{bmatrix}
       a & b \\
       c & d \\
\end{bmatrix}\begin{bmatrix}
       x \\
       y \\
     \end{bmatrix}= \begin{bmatrix}
       ax+by \\
       cx+dy \\
     \end{bmatrix}
\end{align}$ 

$\begin{align}
\begin{bmatrix}
       a & b \\
       c & d \\
\end{bmatrix}
\end{align}$ - the transformation matrix, where $\vec{i}$ and $\vec{j}$ landed

$\begin{align}
\begin{bmatrix}
       x \\
       y \\
\end{bmatrix}
\end{align}$ - the vector to be transformed

**examples**:

- 90 degree rotation ccw $\begin{align}
\begin{bmatrix}
       0 & -1 \\
       1 & 0 \\
\end{bmatrix}
\end{align}$, $\begin{align}
\vec{i}: \begin{bmatrix}
       1 \\
       0 \\
     \end{bmatrix} \rightarrow \begin{bmatrix}
       0 \\
       1 \\
     \end{bmatrix}
\end{align}$, $\begin{align}
\vec{j}: \begin{bmatrix}
       0 \\
       1 \\
     \end{bmatrix} \rightarrow \begin{bmatrix}
       -1 \\
       0 \\
     \end{bmatrix}
\end{align}$

- shear $\begin{align}
\begin{bmatrix}
       1 & 1 \\
       0 & 1 \\
\end{bmatrix}
\end{align}$

**every time you see a matrix, you can interpret it as a certain transformation of space**

**when a matrix multiplies a vector, it means to apply the transformation to the vector**

#### matrix multiplication

**composition** - first rotation then shear

finally $\vec{i}$ and $\vec{j}$ landed in $\begin{align}
\begin{bmatrix}
       1  \\
       1  \\
\end{bmatrix}
\end{align}$ and $\begin{align}
\begin{bmatrix}
       -1  \\
       0  \\
\end{bmatrix}
\end{align}$

this gives a matrix $\begin{align}
\begin{bmatrix}
       1 & -1 \\
       1 & 0 \\
\end{bmatrix}
\end{align}$ called **composition matrix**

if to apply shear + rotation to a vector:

$\begin{align}
\begin{bmatrix}
       1 & 1 \\
       0 & 1 \\
\end{bmatrix}\begin{bmatrix}
       0 & -1 \\
       1 & 0 \\
\end{bmatrix}\begin{bmatrix}
       x \\
       y \\
\end{bmatrix}=\begin{bmatrix}
       1 & -1 \\
       1 & 0 \\
\end{bmatrix}\begin{bmatrix}
       x \\
       y \\
\end{bmatrix}
\end{align}$

in general

$\begin{align}
\begin{bmatrix}
       a & b \\
       c & d \\
\end{bmatrix}\begin{bmatrix}
       e & f \\
       g & h \\
\end{bmatrix}=\begin{bmatrix}
       ae+bg & af+bh \\
       ce+dg & ef+dh \\
\end{bmatrix}
\end{align}$

in this case

$\begin{align}
\vec{i} \rightarrow \begin{bmatrix}
       e \\
       g \\
\end{bmatrix}, \vec{j} \rightarrow \begin{bmatrix}
       f \\
       h \\
\end{bmatrix}
\end{align}$

**order matters**:

$M_1M_2 \neq M_2 M_1$

**determinant**: a special scaling factor that changes the area of two vectors after a linear transformation

e.g. original area of $\begin{align}
\vec{i}=\begin{bmatrix}
       1 \\
       0 \\
\end{bmatrix}\end{align}$ and $\begin{align}\vec{j}=\begin{bmatrix}
       0 \\
       1 \\
\end{bmatrix}\end{align}$ is $1A$

but after the transformation to $\begin{align}
\begin{bmatrix}
       -1 & 1 \\
       -1 & -1 \\
\end{bmatrix}
\end{align}$, the area becomes $2A$

**whenever the orientation of space is inverted, the determinant is negative**

let $det \triangleq |\cdot|$ represent the **determinant**

$\begin{align}
\begin{vmatrix}
a & b \\
c & d \\
\end{vmatrix}=ad-bc
\end{align}$

$\begin{align}
\begin{vmatrix}
1 & 0 & 1  \\
0.5 & 1 & 1.5 \\
1 & 0 & 1
\end{vmatrix}=0
\end{align}$, because the columns are linearly dependent, then the 3d unit vectors will become 2d

some properties:

- $det(M_1 M_2)=det(M_1)det(M_2)$

- when $det(M)\neq 0$ same dimension

meaning space does not get squished into a zero area region

- when $det(M)=0$ lowerizes the dimension

**linear system of equations**: e.g.

$2x+5y+3z=-3$

$4x+0y+8z=0$

$1x+3y+0z=2$

can be turned into

$\begin{align}
\begin{bmatrix}
2 & 5 & 3 \\
4 & 0 & 8 \\
1 & 3 & 0 \\
\end{bmatrix}\begin{bmatrix}
x \\
y \\
z \\
\end{bmatrix}=\begin{bmatrix}
-3 \\
0 \\
2 \\
\end{bmatrix}
\end{align}$

in short $A\vec{x}=\vec{v}$: after a linear transformation $\vec{x}$ becomes $\vec{v}$

**inverse**: $A^{-1}$ do the inverse transformation

e.g. inverse of 90 degree ccw $\begin{align}
\begin{bmatrix}
0 & -1  \\
1 & 0 \\
\end{bmatrix}
\end{align}$ is 90 degree cw $\begin{align}
\begin{bmatrix}
0 & 1  \\
-1 & 0 \\
\end{bmatrix}
\end{align}$

**important property**: 

- $A^{-1}A=I$

so that $A\vec{x}=\vec{v}$ => $\vec{x}=A^{-1}\vec{v}$

- if $det(A)\neq 0$, $A^{-1}$ exists

- if $det(A)=0$, cannot do inverse, since a line can not be inversed into a plane

**rank**: the number of dimension in the output of a transformation

- **rank 1**: the output of a transformation is a line

- **rank 2**: the output of a transformation is 2d plane

- **full rank**: the rank as high as it can be

**column space of a matrix**: the set of all possible outputs for the matrix, no matter a line, a plane a 3d space, namely the span of columns of the matrix

- the zero vector is always in the column space, since linear transformation must keep the origin fixed in place

- for a full rank transformation, the only vector that lands at the origin is the zero vector

- for matrices not full rank, which squish to a smaller dimension, you can have a whole bunch of vectors that land on zero

**null space**: the set of vectors that lands on the origin, the space of all vectors that become null since they landed on the origin

e.g. $\begin{align}
A\vec{x}=\begin{bmatrix}
0  \\
0 \\
\end{bmatrix}
\end{align}$, the null space gives you all of the possible solutions to the equation

in summary:

- linear system equations can have some kind of linear transformation associated with it

- when the transformation has an inverse, you can use the inverse to solve the system equations

- column space lets us understand when a solution even exists

- null space helps us to understand what the set of all possible solutions can look like

**nonsquare matrix**:

$\begin{align}
\begin{bmatrix}
3 & 1 \\
4 & 1 \\
2 & 10 \\
\end{bmatrix}
\end{align}$ transforms 2d space onto 3d

$\begin{align}
\begin{bmatrix}
3 & 1 & 4\\
1 & 5 & 9 \\
\end{bmatrix}
\end{align}$ transforms 3d space onto 2d

**dot product**: a vector projects to another, and multiply up the length of the projections

$\begin{align}
\begin{bmatrix}
u_x & u_y \\
\end{bmatrix}\begin{bmatrix}
x \\
y \\
\end{bmatrix}=u_x x+u_y y
\end{align}$ 

**duality**: natural but surprising correspondence between two types of mathematical thing

- the **dual** of a vector is the linear transformation that it encodes

- the **dual** of a linear transformation from space to 1d is a certain vector in that space

**outer product**: $\vec{v} \times \vec{w}$ = a new vector with the length of area of the parallelogram

$\begin{align}
\vec{v} \times \vec{w}=\begin{bmatrix}
v_1 \\
v_2 \\
v_3 \\
\end{bmatrix} \times \begin{bmatrix}
w_1 & w_2 & w_3 \\
\end{bmatrix}=\begin{vmatrix}
\vec{i} & v_1 & w_1  \\
\vec{j} & v_2 & w_2  \\
\vec{k} & v_3 & w_3  \\
\end{vmatrix} = \vec{i}(v_2w_3-v_3w_2)+\vec{j}(v_3w_1-v_1w_3)+\vec{k}(v_1w_2-v_2w_1)\end{align}$

- $\vec{v} \times \vec{w} = - \vec{w} \times \vec{v}$: order matters

- length of $\begin{align}
\vec{v} \times \vec{w}=\begin{vmatrix}
3 & 2  \\
1 & -1  \\
\end{vmatrix}
\end{align}$ where $\begin{align}
\vec{v}=\begin{bmatrix}
3  \\
1  \\
\end{bmatrix}
\end{align}$ and $\begin{align}
\vec{w}=\begin{bmatrix}
2  \\
-1  \\
\end{bmatrix}
\end{align}$

recall the determinant is all about measuring how areas change due to a transformation

- direction of $\vec{v} \times \vec{w}$ follows the right hand rules

**summary**:

- inner product: lowerize the dimension from 2 vectors into a number

- outer product: expand the dimension from 2 vectors into a matrix

#### change of basis

in our space: 

$\begin{align}
\vec{i}=\begin{bmatrix}
1  \\
0  \\
\end{bmatrix}
\end{align}$, $\begin{align}
\vec{j}=\begin{bmatrix}
0  \\
1  \\
\end{bmatrix}
\end{align}$, aka $\begin{align}
I=\begin{bmatrix}
1 & 0  \\
0 & 1  \\
\end{bmatrix}
\end{align}$


in Jennifer's space:

$\begin{align}
\vec{i}=\begin{bmatrix}
2  \\
1  \\
\end{bmatrix}
\end{align}$, $\begin{align}
\vec{j}=\begin{bmatrix}
-1  \\
1  \\
\end{bmatrix}
\end{align}$, aka $\begin{align}
A=\begin{bmatrix}
2 & -1  \\
1 & 1  \\
\end{bmatrix}
\end{align}$

if in our space we do a 90 degree ccw, we have the transformation matrix as $\begin{align}
M=\begin{bmatrix}
0 & -1  \\
1 & 0  \\
\end{bmatrix}
\end{align}$

what's the transformation matrix in Jennifer's space?

starting from any vector in J's space say $\begin{align}
\begin{bmatrix}
-1  \\
2  \\
\end{bmatrix}
\end{align}$

1. translate it to our space

$\begin{align}
A\begin{bmatrix}
-1  \\
2  \\
\end{bmatrix}
\end{align}$

2. apply the transformation matrix in our space

$\begin{align}
MA\begin{bmatrix}
-1  \\
2  \\
\end{bmatrix}
\end{align}$

3. translate it back to J's space

$\begin{align}
A^{-1}MA\begin{bmatrix}
-1  \\
2  \\
\end{bmatrix}
\end{align}$

**an expression like $A^{-1}MA$ suggests a mathematical sort of empathy, that middle matrix represents a transformation of some kind, and the outer two matrices represent the empathy: the shift in perspective**

**$A^{-1}MA$ represents the same transformation as $M$ but with a shifted view (change of basis)**

#### eigenvalue and eigenvector

when doing a linear transformation say $\begin{align}
\begin{bmatrix}
3 & 1  \\
0 & 2  \\
\end{bmatrix}
\end{align}$, most of the vectors will get knocked off their span during the transformation

span: the line passing through the original vector, or all the vectors in the same line

knocked off: the transformed vector shifted from the span of original vector

**however**, some special vectors do remain on their own span

e.g. $\begin{align}
\begin{bmatrix}
1  \\
0  \\
\end{bmatrix}
\end{align}$,  $\begin{align}
\begin{bmatrix}
-1  \\
1  \\
\end{bmatrix}
\end{align}$

$\begin{align}
\begin{bmatrix}
1  \\
0  \\
\end{bmatrix}
\end{align}$ stretched by 3

$\begin{align}
\begin{bmatrix}
-1  \\
1  \\
\end{bmatrix}
\end{align}$ streched by 2

these are the **eigenvectors**, and 3,2 are the **eigenvalues** of the transformation

**eigen vectors are the spans do not change with the transformation**, for example can be regarded as fixed axes

**def**:

$A\vec{v}=\lambda \vec{v}$

the matrix vector product $A$ times $\vec{v}$ gives the same result as just scaling the eigenvector $\vec{v}$ by some value $\lambda$

$A\vec{v}=\lambda \vec{v}$

=> $A\vec{v}=\lambda I \vec{v}$

=> $(A-\lambda I)\vec{v}=\vec{0}$

=> $det(A-\lambda I)=0$, $det(A)=0$ squishes down dimension

e.g. find the eigenvectors and eigenvalues of $\begin{align}
\begin{bmatrix}
3 & 1  \\
0 & 2  \\
\end{bmatrix}
\end{align}$

$\begin{align}
\begin{vmatrix}
3-\lambda & 1  \\
0 & 2-\lambda  \\
\end{vmatrix}=0
\end{align}$

=> $\lambda=3$ or $2$

put back $\lambda$ to the matrix

$\begin{align}
\begin{bmatrix}
3-2 & 1  \\
0 & 2-2  \\
\end{bmatrix}\begin{bmatrix}
x \\
y \\
\end{bmatrix}=
\begin{bmatrix}
0 \\
0 \\
\end{bmatrix}
\end{align}$

solutions:

$\begin{align}
\begin{bmatrix}
-1 \\
1  \\
\end{bmatrix}
\end{align}$, $\begin{align}
\begin{bmatrix}
1 \\
0  \\
\end{bmatrix}
\end{align}$

some transformation doesn't have eigenvector, i.e. $\begin{align}
\begin{bmatrix}
0 & -1 \\
1 & 0 \\
\end{bmatrix}
\end{align}$

**note: in a diagonal matrix, all the diagonal entries of this matrix is eigenvalues**

**eigenbasis**: use eigenvectors as a new basis or coordinate

use eigenvectors $\begin{align}
A=\begin{bmatrix}
1 & -1 \\
0 & 1  \\
\end{bmatrix}
\end{align}$ of the matrix $\begin{align}
M=\begin{bmatrix}
3 & 1 \\
0 & 2  \\
\end{bmatrix}
\end{align}$ as a new basis vectors

you can do $A^{-1}MA=\begin{align}
\begin{bmatrix}
3 & 0 \\
0 & 2  \\
\end{bmatrix}
\end{align}$ to have a new coordinate which is guaranteed to be diagonal

you can do operation or calculation in this new corrdinate and then transform back to the original coordinate for simplicity

**some properties**:

1. $\begin{align}
\frac{1}{2} tr\left(\begin{bmatrix}
a & b \\
c & d  \\
\end{bmatrix}\right)=\frac{a+d}{2}=\frac{\lambda_1+\lambda_2}{2}=m
\end{align}$

mean of $a,d$ = mean of $\lambda_1, \lambda_2$

2. $\begin{align}
\begin{vmatrix}
a & b \\
c & d  \\
\end{vmatrix}=ad-bc=\lambda_1 \lambda_2 = p
\end{align}$

$det(A)$ = product of $\lambda_1, \lambda_2$

3. a quick way to solve $\lambda_1, \lambda_2=m \pm \sqrt{m^2-p}$ which has a relation with roots of quadratic polynomial

```python
plt.figure(figsize=(6,6))
plt.arrow(0,0,1,2,color='orange',linewidth=3,head_width=0.2,length_includes_head=True,label='[1,2]')
plt.arrow(0,0,1,0,color='b',linewidth=3,head_width=0.2,length_includes_head=True,label='i=[1,0]')
plt.arrow(0,0,0,1,color='g',linewidth=3,head_width=0.2,length_includes_head=True,label='j=[0,1]')
plt.arrow(0,0,3,0,color='b',linestyle='--',linewidth=3,head_width=0.2,length_includes_head=True,label='i new=[3,0]')
plt.arrow(0,0,1,2,color='g',linestyle='--',linewidth=3,head_width=0.2,length_includes_head=True,label='j new=[1,2]',alpha=0.5)
plt.arrow(0,0,5,4,color='orange',linestyle='--',linewidth=3,head_width=0.2,length_includes_head=True,label='new=[5,4]')
plt.plot(0,0,'ro',markersize=12,label='origin')
plt.plot(np.linspace(-3,10),2*np.linspace(-3,10),'r',label='original span',alpha=0.5)
plt.plot(np.linspace(-3,10),(4./5)*np.linspace(-3,10),'r',label='new span',alpha=0.5)
plt.axhline(y=0,color='b',label='unchanged span [1,0]',alpha=0.5)
plt.plot(np.linspace(-3,10),-np.linspace(-3,10),'b',label='unchanged span [-1,1]',alpha=0.5)
plt.legend(loc='center left',fontsize=14,bbox_to_anchor=(1, 0.5))
plt.xlim([-3,10])
plt.ylim([-3,10])
plt.xticks(np.arange(-3,10))
plt.yticks(np.arange(-3,10))
plt.grid()
plt.gca().set_aspect("equal")
```
